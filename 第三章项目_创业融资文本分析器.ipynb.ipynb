```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.stem import PorterStemmer, WordNetLemmatizer
import os

# 步骤1. 读取本地txt新闻
txt_path = 'startup_funding.txt'
with open(txt_path, 'r', encoding='utf-8') as f:
    raw = f.read()

# 步骤2. 分词
tokens = word_tokenize(raw)
words = [w.lower() for w in tokens if w.isalpha()]

# 步骤3. 高频词统计
fdist = FreqDist(words)
print("——高频词统计（前10）——")
for word, count in fdist.most_common(10):
    print(f"{word:<12s} : {count}")

# 步骤4. 词干提取与原型化
porter = PorterStemmer()
wnl = WordNetLemmatizer()
sample_words = list(fdist)[:10]
print("\n——词干提取和原型化展示——")
for w in sample_words:
    print(f"Word: {w:12s} | Stem: {porter.stem(w):12s} | Lemma: {wnl.lemmatize(w)}")

# 步骤5. 关键词上下文查询
target = "funding"
text = nltk.Text(words)
print(f"\n——‘{target}’的左右上下文片段——")
text.concordance(target)

# 步骤6. 统计“ing”结尾和“capital”相关词
ing_words = [w for w in words if w.endswith('ing')]
capital_related = [w for w in words if 'capital' in w]
print(f"\n含‘ing’结尾的词: {set(ing_words)}")
print(f"含‘capital’的词: {set(capital_related)}")

# 步骤7. 简单情感正负面字计数
pos_words = {'success','scalable','growth','improve','adaptability','sustainable'}
neg_words = {'uncertainties','competitive'}
pos = sum(fdist[w] for w in pos_words if w in fdist)
neg = sum(fdist[w] for w in neg_words if w in fdist)
print(f"\n文本正面词数量: {pos}，负面词数量: {neg}")

# NLTK数据提示
print("\n如果首次运行报错，请先执行:")
print("  nltk.download('punkt')\n  nltk.download('wordnet')")
```
