```python
for fileid in gutenberg.fileids():
    num_chars=len(gutenberg.raw(fileid))#这本书总共有多少个【字符】（字母、标点、空格都算）
    num_words=len(gutenberg.words(fileid))#这本书总共有多少个【单词】
    num_sents=len(gutenberg.sents(fileid))#这本书总共有多少个【句子】
    num_vocab=len(set([w.lower() for w in gutenberg.words(fileid)]))#这本书有多少个【不重复单词】（去重后的词汇量）
    print(int(num_chars/num_words),int(num_words/num_sents),int(num_words/num_vocab),fileid)#平均词长 平均句子长度 每个词平均出现次数
```

```python
import nltk
from nltk.corpus import inaugural

# 正确创建条件频率分布
cfd = nltk.ConditionalFreqDist()

# 循环遍历每一篇就职演说
for fileid in inaugural.fileids():
    year = fileid[:4]  # 提取年份
    
    # 遍历每一个单词
    for w in inaugural.words(fileid):
        word = w.lower()
        
        # 判断是否以 America 或 citizen 开头
        for target in ['america', 'citizen']:
            if word.startswith(target):
                cfd[target][year] += 1  # 正确计数

# 最后画图
cfd.plot()
```

```python
import nltk
from nltk.corpus import udhr
languages = ['English', 'German_Deutsch', 'Chickasaw']

# 创建条件频率分布（空的，用来存数据）
cfd = nltk.ConditionalFreqDist()

# 循环：每种语言 → 每个单词 → 统计单词长度
for lang in languages:
    # 读取对应语言的《世界人权宣言》单词
    for word in udhr.words(lang + '-Latin1'):
        cfd[lang][len(word)] += 1

# 画出累积分布图
cfd.plot(cumulative=True)
cfd.tabulate(conditions=['English','German_Deutsch'],samples=range(10),cumulative=True)
#samples=range(10),样本是：单词长度 0~9,也就是看：1 个字母、2 个字母…9 个字母 的单词
```

```python
import nltk
from nltk.corpus import PlaintextCorpusReader#导入你自己的语料库
import matplotlib.pyplot as plt

# ----------------------
# 1. 导入你桌面的语料库
# ----------------------
corpus_root = r"C:\Users\78738\Desktop\my_corpus"  # 你建的文件夹
my_corpus = PlaintextCorpusReader(corpus_root, ".*\.txt")

print("✅ 你的语料库文件：")
print(my_corpus.fileids())

# ----------------------
# 2. 基础统计分析
# ----------------------
words = my_corpus.words()
print("\n总单词数：", len(words))
print("不重复单词数：", len(set(words)))

# ----------------------
# 3. 统计高频词（最常用20个）
# ----------------------
fdist = nltk.FreqDist(w.lower() for w in words)
print("\n🔝 高频词前20：")
for word, count in fdist.most_common(20):
    print(f"{word}: {count}")

# ----------------------
# 4. 画图1：高频词频率图
# ----------------------
plt.figure(figsize=(10, 5))
fdist.plot(20, title="我的语料库 - 高频词分布")

# ----------------------
# 5. 画图2：条件频率分布（对比3个文本）
# ----------------------
cfd = nltk.ConditionalFreqDist()
for fileid in my_corpus.fileids():
    for word in my_corpus.words(fileid):
        cfd[fileid][len(word)] += 1

plt.figure(figsize=(10, 5))
cfd.plot(title="三个文本的单词长度对比")

# ----------------------
# 6. 画图3：单词长度分布
# ----------------------
len_dist = nltk.FreqDist(len(w) for w in words)
plt.figure(figsize=(10, 5))
len_dist.plot(title="单词长度分布")

plt.show()
```

generate_model：生成文本的函数
cfdist：传入 “单词概率表”（知道每个单词后面最可能跟什么）
word：从哪个单词开始生成
num=15：默认生成 15 个单词
print(word)：输出当前单词
word = cfdist[word].max()
→ 核心逻辑：
查概率表 → 找到当前词后面出现次数最多的词 → 把它设为下一个词

```python
def generate_model(cfdist,word,num=15):
    for i in range(num):
        print(word)
        word=cfdist[word].max()
text=nltk.corpus.genesis.words('english-kjv.txt')
bigrams=nltk.bigrams(text)
cfd=nltk.ConditionalFreqDist(bigrams)
```

```python
print(cfd['living']),generate_model(cfd,'living')
```

```python
def unusual_words(text):
    text_vocab=set(w.lower() for w in text if w.isalpha())
    english_vocab=set(w.lower() for w in nltk.corpus.words.words())
    unusual=text_vocab.difference(english_vocab)#文本单词 - 标准词典单词 = 奇怪/生僻/专有名词
    return sorted(unusual)
unusual_words(nltk.corpus.gutenberg.words('austen-sense.txt'))
```

```python
#计算文本中不含停用词中词所占比例
def content_fraction(text):
    stopwords=nltk.corpus.stopwords.words('english')
    content=[w for w in text if w.lower() not in stopwords]
    return len(content)/len(text)
content_fraction(nltk.corpus.reuters.words())
```
