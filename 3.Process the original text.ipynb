```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import nltk
nltk.download('punkt')

# 1. 爬取人民网
url = "https://www.people.com.cn/"
html = urlopen(url).read().decode('utf-8')

# 2. 清理 HTML，只留正文（关键！）
soup = BeautifulSoup(html, 'html.parser')
raw = soup.get_text()  # 只有这行能去掉所有标签！

# 3. 分词
tokens = nltk.word_tokenize(raw)
text = nltk.Text(tokens)

# 4. 搜索
print("=== 搜索：人民 ===")
text.concordance("title")
```

```python
def stem(word):
    regexp=r'^(.*?)(ing|ly|ed|ious|ies|ive|es|s|ment)?$'#. = 任意字符 * = 任意多个 ? = 非贪婪（尽量少匹配，把后面留给后缀）
    stem,suffix=re.findall(regexp,word)[0]#切成词根和后缀
    return stem
raw="Investment is the key to financial growth. People save money and put it into stocks, funds, or real estate to gain more wealth over time. Understanding risk and return helps everyone make better decisions. A balanced portfolio can protect your money and increase value steadily. Learning basic finance is useful for daily life and long-term planning."
tokens=nltk.word_tokenize(raw)
[stem(t) for t in tokens] 
```

```python
def tabulate(cfdist, words, categories):
    print('%-16s' % 'Category')
    for word in words:
        print('%6s' % word)
    print()
    for category in categories:
        print('%-16s' % category)
        for word in words:
            print('%6d' % cfdist[category][word])
        print()

from nltk.corpus import brown
import nltk

cfd = nltk.ConditionalFreqDist(
    (genre, word)
    for genre in brown.categories()
    for word in brown.words(categories=genre)
)

genres = ['news','religion','hobbies','science_fiction','romance']
modals = ['can','could','may','might','must','will']

tabulate(cfd, modals, genres)
```
